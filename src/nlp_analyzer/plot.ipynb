# Create some exporatory plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
amlo_score = pd.read_csv(
    "C:/Users/fdmol/Desktop/AMLO-NLP/src/data/nnet_agressivity_scores.csv"
)
amlo_score.sort_values(by="id", inplace=True)

amlo_score["date"] = pd.to_datetime(amlo_score["id"], format="%Y%m%d")

# amlo_score['week'] = amlo_score['date'].dt.day_name()
# amlo_score['month'] = amlo_score['date'].dt.month_name()
# amlo_score['year'] = amlo_score['date'].dt.year

# Filter by date
# amlo_score = amlo_score[(amlo_score["date"] >= "2024-01-01")]

# amlo

# Lineplot score vs date
sns.set_theme()
plt.figure(figsize=(15, 10))
sns.lineplot(x="date", y="score", data=amlo_score)
plt.title("AMLO agressivity score")
plt.show()
```

```python
# Lineplot num_words vs date
plt.figure(figsize=(12, 10))
sns.lineplot(x="date", y="num_words", data=amlo_score)
plt.title("AMLO number of words")
plt.show()
```
